```python
import tkinter as tk
from tkinter import messagebox

# Game data
player = {"health": 100, "inventory": []}

rooms = {
    "Entrance": {"description": "You are at the dungeon entrance.", "north": "Hall"},
    "Hall": {"description": "A dark hall with a shiny sword.", "south": "Entrance", "east": "Treasure Room", "item": "Sword"},
    "Treasure Room": {"description": "A room with a sleeping dragon!", "west": "Hall", "monster": "Dragon"},
}

current_room = "Entrance"

# Functions
def show_status():
    status_text.set(f"Room: {current_room}\nHealth: {player['health']}\nInventory: {player['inventory']}")
    room_text.set(rooms[current_room]["description"])

def move(direction):
    global current_room
    if direction in rooms[current_room]:
        current_room = rooms[current_room][direction]
        if "monster" in rooms[current_room]:
            fight(rooms[current_room]["monster"])
        show_status()
    else:
        messagebox.showinfo("Can't move", "You can't go that way!")

def pick_item():
    room = rooms[current_room]
    if "item" in room:
        player["inventory"].append(room["item"])
        messagebox.showinfo("Item picked", f"You picked up {room['item']}!")
        del room["item"]
        show_status()
    else:
        messagebox.showinfo("No item", "There is nothing to pick up here.")

def fight(monster):
    if "Sword" in player["inventory"]:
        messagebox.showinfo("Fight", f"You fought the {monster} and won!")
    else:
        player["health"] -= 50
        messagebox.showinfo("Fight", f"You fought the {monster} and got hurt! Health -50")
    show_status()

# GUI setup
root = tk.Tk()
root.title("Mini Adventure Game")

room_text = tk.StringVar()
status_text = tk.StringVar()

tk.Label(root, textvariable=room_text, wraplength=300, height=5).pack(pady=10)
tk.Label(root, textvariable=status_text).pack(pady=10)

frame = tk.Frame(root)
frame.pack(pady=10)

tk.Button(frame, text="North", command=lambda: move("north")).grid(row=0, column=1)
tk.Button(frame, text="South", command=lambda: move("south")).grid(row=2, column=1)
tk.Button(frame, text="East", command=lambda: move("east")).grid(row=1, column=2)
tk.Button(frame, text="West", command=lambda: move("west")).grid(row=1, column=0)
tk.Button(root, text="Pick Item", command=pick_item).pack(pady=5)

show_status()
root.mainloop()
```
