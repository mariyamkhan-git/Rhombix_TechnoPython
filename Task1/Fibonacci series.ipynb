```python
def fibonacci(number):
    if number <= 0:
        print("No Fibonacci numbers to display.")
        print("The total sum of Fibonacci series is: 0")
        return

    a = 0
    b = 1
    total = 0

    for i in range(number):
        print(a, end=" ")
        total += a

        temp = a + b   # store next Fibonacci number
        a = b          # move b into a
        b = temp       # move temp into b

    print("\nThe total sum of Fibonacci series is:", total)
```

```python
print("Please enter a number of your choice:")
num=int(input()) 
  
fibonacci(num)
```
